```python
url = "http://lyrics.wikia.com/api.php?action=lyrics&artist={0}".format("Eminem")
soup = BeautifulSoup(requests.get(url).text)
print(soup.find("ul",{'class':'albums'}))

for al in soup.find("ul",{'class':'albums'}):
    if al.lower() == "Infinite":
        currentAlbum = al
        break
songs =[Track(song.text,album,album.artist()) for song in currentAlbum.findNext('songs').findAll('item')]
print(songs)
```
